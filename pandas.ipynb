```python
data={
    "Name":["Khan","Ainan","Ahmed","Ayan","Shaikh","Faizan","Sheikh","Ansari"],
    "Age":[19,21,22,45,75,35,34,87],
    "Salary":[10000,20000,40000,45000,23000,21000,86000,65000],
    "Performance Score":[6,3,7,5,2,9,5,1]
}
df=pd.DataFrame(data)
print(df)
```

```python
print(df.describe())
```

```python
print(df.shape)
print(df.columns)
```

```python
print(df["Name"])  #Select single column
```

```python
print(df[["Name","Salary"]])   #Select multiple columns
```

```python
highest_salarries=df[df["Salary"]>25000]  #select on basis of single condition
print("Highest Salary")
print(highest_salarries)
```

```python
filter=df[(df["Salary"]>40000) & (df["Age"]>40)]    #select on basis of multiple condition
print(filter)
```

```python
df["Bonus"]=df["Salary"]*0.1  #Adding a column in dataset
print(df)
```

```python
df.insert(0,"Emp_id",[1,2,3,4,5,6,7,8])  #inser column at certain position
print(df)
```

```python
df.loc[1,"Salary"]=70000  #update single value
print(df)
highest_salary=df[df["Salary"]>69000]
print("Highest Salary")
print(highest_salary)
```

```python
df["Salary"]=df["Salary"]*1.05 #updating multiple values, increasing by 5%
print(df)
```

```python
df.drop(columns=["Performance Score","Age"], inplace=True)  #deleting column
print(df)
```
